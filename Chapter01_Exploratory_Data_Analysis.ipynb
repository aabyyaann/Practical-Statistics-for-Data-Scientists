# Import Library

Library yang digunakan:

- NumPy
- Pandas
- Matplotlib
- Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.set_printoptions(precision=4, suppress=True)

sns.set_style("whitegrid")
```

# Dataset Contoh

Pada chapter ini digunakan dataset sederhana berupa populasi beberapa negara bagian (state).

```python
data = pd.DataFrame({
    "State":["A","B","C","D","E"],
    "Population":[500000,700000,300000,900000,800000]
})

data
```

# 1. Mean

Mean adalah rata-rata aritmatika dari seluruh data.

\[
Mean=\frac{\sum x_i}{n}
\]

```python
mean_pop = data["Population"].mean()

print(mean_pop)
```

# Interpretasi

Mean menunjukkan nilai pusat dari data populasi.

# 2. Median

Median adalah nilai tengah setelah data diurutkan.

```python
median_pop = data["Population"].median()

print(median_pop)
```

# Mengapa Median Penting?

Median lebih tahan terhadap outlier dibanding mean.

# 3. Weighted Mean

Weighted Mean memberikan bobot berbeda pada setiap observasi.

```python
values = np.array([10,20,30,40])

weights = np.array([1,2,3,4])

weighted_mean = np.average(
    values,
    weights=weights
)

print(weighted_mean)
```

# 4. Trimmed Mean

Trimmed Mean menghapus sebagian data ekstrem sebelum menghitung rata-rata.

```python
from scipy.stats import trim_mean

sample = [10,12,15,17,18,20,21,200]

trim_mean(sample,0.1)
```

# Interpretasi

Trimmed Mean membantu mengurangi pengaruh outlier.

# 5. Variance

Variance mengukur penyebaran data terhadap mean.

\[
Var(X)=\frac{\sum(x_i-\bar{x})^2}{n-1}
\]

```python
variance = data["Population"].var()

print(variance)
```

# 6. Standard Deviation

Standard deviation adalah akar dari variance.

```python
std_dev = data["Population"].std()

print(std_dev)
```

# Interpretasi

Semakin besar standard deviation, semakin menyebar data.

# 7. Percentiles

Percentile membagi data menjadi 100 bagian.

```python
np.percentile(
    data["Population"],
    [25,50,75]
)
```

# Quartiles

Q1 = 25th percentile

Q2 = Median

Q3 = 75th percentile

# 8. Interquartile Range (IQR)

\[
IQR = Q3 - Q1
\]

```python
q1 = np.percentile(
    data["Population"],
    25
)

q3 = np.percentile(
    data["Population"],
    75
)

iqr = q3 - q1

print(iqr)
```

# Mengapa IQR Penting?

IQR sering digunakan untuk mendeteksi outlier.

# 9. Boxplot

Boxplot membantu visualisasi distribusi data.

```python
plt.figure(figsize=(6,4))

plt.boxplot(
    data["Population"]
)

plt.title("Population Boxplot")

plt.show()
```

# Interpretasi Boxplot

Boxplot menunjukkan:

- Median
- Quartiles
- Outliers

# 10. Histogram

Histogram menunjukkan distribusi frekuensi data.

```python
plt.figure(figsize=(6,4))

plt.hist(
    data["Population"],
    bins=5
)

plt.title("Population Histogram")

plt.show()
```

# Histogram Interpretation

Histogram membantu melihat bentuk distribusi data.

# 11. Density Plot

Density plot memberikan estimasi distribusi probabilitas data.

```python
plt.figure(figsize=(6,4))

sns.kdeplot(
    data["Population"],
    fill=True
)

plt.show()
```

# 12. Categorical Data

```python
categories = pd.Series([
    "A",
    "B",
    "A",
    "C",
    "A",
    "B"
])

categories.value_counts()
```

# Frekuensi Kategori

```python
categories.value_counts().plot(
    kind='bar'
)

plt.show()
```

# 13. Correlation

Correlation mengukur hubungan linear antar variabel.

Nilai berada pada rentang:

\[
-1 \le r \le 1
\]

```python
corr_df = pd.DataFrame({
    "X":[1,2,3,4,5],
    "Y":[2,4,6,8,10]
})

corr_df.corr()
```

# Interpretasi Correlation

- +1 = hubungan positif sempurna
- 0 = tidak ada hubungan linear
- -1 = hubungan negatif sempurna

# 14. Scatter Plot

```python
plt.figure(figsize=(6,4))

plt.scatter(
    corr_df["X"],
    corr_df["Y"]
)

plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

# Interpretasi Scatter Plot

Scatter plot membantu mengidentifikasi:

- hubungan linear
- outlier
- pola data
